```python
import math
```

```python
q96 = 2**96
```

```python
def price_to_sqrtp(p):
    return int(math.sqrt(p) * q96)


def sqrtp_to_price(sqrtp):
    return (sqrtp / q96)**2
```

```python
# PARAMETERS
ethPriceInUSD = 4300

# SUP PRICE TARGET
supPriceInUSD = 0.045
supPriceInETH = supPriceInUSD / ethPriceInUSD
ethPriceInSUP = 1 / supPriceInETH

print("SUP PRICE TARGET :")
print("--------------------------------")
print(supPriceInUSD, "USD/SUP")
print(supPriceInETH, "ETH/SUP")
print(ethPriceInSUP, "SUP/ETH")
```

```python
print("PRICE TO USE IF SUP IS TOKEN 0 IN ETH/SUP POOL")

supPriceInETHQ96 = price_to_sqrtp(supPriceInETH)
print(supPriceInETHQ96)
sqrtp_to_price(supPriceInETHQ96)
```

```python
print("PRICE TO USE IF SUP IS TOKEN 1 IN ETH/SUP POOL")
ethPriceInSUPQ96 = price_to_sqrtp(ethPriceInSUP)
print(ethPriceInSUPQ96)
sqrtp_to_price(ethPriceInSUPQ96)
```
